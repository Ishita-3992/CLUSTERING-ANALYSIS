#DATA PREPROCESSING

NORMALISATION

```python
setup(data = p, normalize = True, normalize_method = 'zscore', verbose=False)
x = create_model('meanshift')
```

```python
print("For Cluster = 3")
x = create_model('meanshift', num_clusters = 3)

print("For Cluster = 4")
x = create_model('meanshift', num_clusters = 4)

print("For Cluster = 5")
x = create_model('meanshift', num_clusters = 5)
```

#TRANSFORMATION

```python
setup(data = p, transformation = True, transformation_method = 'yeo-johnson', verbose=False)
x = create_model('meanshift')
```

```python
print("For Cluster = 3")
x = create_model('meanshift', num_clusters = 3)

print("For Cluster = 4")
x = create_model('meanshift', num_clusters = 4)

print("For Cluster = 5")
x = create_model('meanshift', num_clusters = 5)
```

#PCA

```python
setup(data = p, pca = True, pca_method = 'linear', verbose=False)
x = create_model('meanshift')
```

```python
print("For Cluster = 3")
x = create_model('meanshift', num_clusters = 3)

print("For Cluster = 4")
x = create_model('meanshift', num_clusters = 4)

print("For Cluster = 5")
x = create_model('meanshift', num_clusters = 5)
```

# "Transformation" + "Normalization"

```python
setup(data = p, transformation = True, normalize = True,
      normalize_method = 'zscore', transformation_method = 'yeo-johnson', verbose=False)
x = create_model('meanshift')
```

```python
print("For Cluster = 3")
x = create_model('meanshift', num_clusters = 3)

print("For Cluster = 4")
x = create_model('meanshift', num_clusters = 4)

print("For Cluster = 5")
x = create_model('meanshift', num_clusters = 5)
```

#"Transformation" + "Normalization" + "PCA"

```python
print("For Cluster = 3")
x = create_model('meanshift', num_clusters = 3)

print("For Cluster = 4")
x = create_model('meanshift', num_clusters = 4)

print("For Cluster = 5")
x = create_model('meanshift', num_clusters = 5)
```
